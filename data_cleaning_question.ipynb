```python
import re
def clean_string(string):
    string=string.replace('-','')
    string=string.replace('..','')
    string=string.replace(';','')
    string=string.replace('.','')
    string=string.replace('i.e.','')
    string=string.replace('e.g.','')
    tmp=string
    # Remove Hashtags
    tmp = re.sub(r'#(\w+)', '', tmp)
    # Remove repeating chars
    tmp = re.sub(r'(.)\1{1,}', r'\1\1', tmp)
    # Remove anything that is not letters
    tmp = re.sub("[^a-zA-Z0-9,/]", " ", tmp)
    # Remove anything that is less than two characters
    tmp = re.sub(r'\b\w{1,2}\b', '', tmp)
    # Remove multiple spaces
    tmp = re.sub(r'\s\s+', ' ', tmp)
    
    tmp = re.sub(r"^\W+", "", tmp)
    
   
    return tmp 
```

```python
for i in range(0,raw.shape[0]):
    raw["RAW DATA"][i]=clean_string(raw["RAW DATA"][i])
    print(raw["RAW DATA"][i])
```
